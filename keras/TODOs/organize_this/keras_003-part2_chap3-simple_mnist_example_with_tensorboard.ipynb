```python
# keras_003-part2_chap3-simple_mnist_example_with_tensorboard
#
# ,      
# Part2.  
# Chapter3.  
# 2018-10-02 ()

# p.51, 2.  

# 0.   
import keras
from keras.utils import np_utils
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Activation
import numpy as np

#1.  
#   
(x_train, y_train), (x_test, y_test) = mnist.load_data()

#   
x_val = x_train[50000:]
y_val = y_train[50000:]
x_train = x_train[:50000]
y_train = y_train[:50000]

#  
x_train = x_train.reshape(50000,784).astype('float32') / 255.0
x_val = x_val.reshape(10000,784).astype('float32') / 255.0
x_test = x_test.reshape(10000,784).astype('float32') / 255.0

#   
# random indexs
np.random.seed(3)
train_rand_idxs = np.random.choice(5000,700)
val_rand_idxs = np.random.choice(10000, 300)

x_train = x_train[ train_rand_idxs ]
y_train = y_train[ train_rand_idxs ]
x_val = x_val[ val_rand_idxs ]
y_val = y_val[ val_rand_idxs ]

#  one-hot encoding 
y_train = np_utils.to_categorical( y_train )
y_val = np_utils.to_categorical( y_val )
y_test = np_utils.to_categorical( y_test )

# 2.  
model = Sequential()
model.add( Dense( units = 2, input_dim = 28*28, activation = 'relu') )
model.add( Dense( units = 10, activation = 'softmax') )

# 3.   
model.compile( loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'] )

# 4.   
tb_hist = keras.callbacks.TensorBoard( log_dir='.graph', histogram_freq=0,)
hist = model.fit( x_train, y_train, epochs = 500, batch_size = 10, validation_data = ( x_val, y_val ), callbacks=[tb_hist])

# 5.  
#        .
# tensorboard --logdir=absolute_path
# 
# (keras) C:\Users\phil4>tensorboard --logdir=C:\Users\phil4\Dropbox\1-today\deel_learning-keras\.graph
# TensorBoard 1.10.0 at http://DESKTOP-G19AR5M:6006 (Press CTRL+C to quit)
# 
#   "http://DESKTOP-G19AR5M:6006" 
```
